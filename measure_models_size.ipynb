```python
from transformers import AutoTokenizer, AutoModelForSequenceClassification
import torch

# Bio_ClinicalBERT (teacher) model size calculation
teacher_model = AutoModelForSequenceClassification.from_pretrained("/content/drive/MyDrive/Colab Notebooks/Postdoc/Bio_ClinicalBERT-fine_tuned/v3-final")
teacher_tokenizer = AutoTokenizer.from_pretrained("emilyalsentzer/Bio_ClinicalBERT")

teacher_path = "/content/drive/MyDrive/Colab Notebooks/Postdoc/Bio_ClinicalBERT-fine_tuned/v3-final/model_size.pt"
torch.save(teacher_model.state_dict(), teacher_path)

import os

def get_file_size(path):
    return os.path.getsize(path) / (1024 ** 2)  # MB

print(f"Original model size: {get_file_size(teacher_path):.2f} MB")
```
